# I. TÌM HIỂU DỮ LIỆU

1. Tên: Là tên của cầu thủ. Đây là trường cơ bản và quan trọng để nhận diện cầu thủ.
2. CLB (Câu lạc bộ): Tên câu lạc bộ mà cầu thủ đang thi đấu hoặc đã thi đấu. CLB giúp xác định đội bóng mà cầu thủ gắn bó.
3. Quốc tịch: Quốc gia mà cầu thủ mang quốc tịch, thể hiện nguồn gốc và quốc gia mà cầu thủ đại diện.
4. Vị trí: Vị trí của cầu thủ trên sân bóng, cho biết vai trò của cầu thủ trong đội bóng.
5. Tuổi: Tuổi của cầu thủ, giúp xác định giai đoạn trong sự nghiệp của họ (trẻ, trung niên, hay sắp giải nghệ).
6. Chiều cao: Chiều cao của cầu thủ, có thể ảnh hưởng đến phong cách thi đấu (chẳng hạn như việc tranh chấp bóng bổng).
7. Cân nặng: Cân nặng của cầu thủ, có thể ảnh hưởng đến tốc độ, sức mạnh và khả năng chịu đựng trong các tình huống thi đấu.
8. Bàn thắng: Số lượng bàn thắng mà cầu thủ ghi được trong một mùa giải hoặc trong sự nghiệp thi đấu. Thường dùng để đánh giá khả năng tấn công của cầu thủ.
9. Kiến tạo: Số lần cầu thủ kiến tạo (tạo ra cơ hội cho đồng đội ghi bàn), thể hiện khả năng chơi bóng nhóm và sáng tạo.
10. Tắc bóng (Tackling): Là hành động cướp bóng từ đối thủ bằng cách sử dụng chân.
11. Đánh chặn (Interception): Là hành động ngừng một đường chuyền của đối thủ và giành lại bóng.
12. Cứu thua (Save): Là hành động thủ môn ngăn bóng vào lưới, bảo vệ khung thành. 
13. Thẻ vàng: Số thẻ vàng mà cầu thủ đã nhận trong các trận đấu. Thẻ vàng thể hiện mức độ kỷ luật và tác động của cầu thủ trong các tình huống tranh chấp.
14. Lạm phát: Đây có thể là thông tin về tỷ lệ lạm phát trong nền kinh tế mà ảnh hưởng đến giá trị chuyển nhượng cầu thủ, hoặc có thể là các vấn đề liên quan đến tài chính trong môi trường bóng đá.
15. Mô tả đặc điểm cá nhân: Miêu tả các đặc điểm về tính cách, phong thái thi đấu, hoặc bất kỳ yếu tố nào khác liên quan đến cá nhân cầu thủ.
16. Phong cách và lối chơi: Thể hiện cách cầu thủ thi đấu trên sân, bao gồm kỹ thuật, chiến thuật, cũng như cách họ tương tác với đồng đội và đối thủ.
17. Điểm mạnh: Những yếu tố mà cầu thủ có thể làm rất tốt hoặc là đặc điểm nổi bật của họ trong thi đấu.
18. Điểm yếu: Những khía cạnh mà cầu thủ còn yếu hoặc cần cải thiện trong thi đấu.
19. Giá: Giá trị chuyển nhượng của cầu thủ, thể hiện sự đánh giá của thị trường và các đội bóng về giá trị tài năng của họ.
20. Nguồn: Chứa ảnh cầu thủ

# III. XÂY DỰNG MÔ HÌNH

```python
# Danh sách từ dừng
stop_words_vn = ["và", "của", "là", "có", "một", "những", "trên", "được", "với", "này", "để"]

def preprocess_text(text):
    # Chuyển về chữ thường và loại bỏ ký tự đặc biệt
    text = re.sub(r'[^\w\s]', '', text.lower())
    # Tách từ và loại bỏ từ dừng
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words_vn]
    return ' '.join(tokens)


# Tạo cột "combined_data" để gom tất cả thông tin liên quan
data['combined_data'] = (
    data['Tên'] + ' ' +
    data['Điểm mạnh'] + ' ' +
     data['Phong cách và lối chơi'] + ' ' +
    data['Mô tả đặc điểm cá nhân'] + ' ' +
    data['Giá'].astype(str)
).apply(preprocess_text)

# Khởi tạo bộ vectorizer TF-IDF và tính toán ma trận TF-IDF cho cột combined_data
tfidf = TfidfVectorizer(stop_words=stop_words_vn)
tfidf_matrix = tfidf.fit_transform(data['combined_data'])

# Hàm gợi ý dựa trên mô tả người dùng
def get_recommendations_by_description(user_input):
    # Chuyển từ khóa tìm kiếm thành TF-IDF vector
    user_tfidf = tfidf.transform([user_input])
    
    # Tính độ tương đồng cosine giữa từ khóa và dữ liệu
    sim_scores = list(enumerate(linear_kernel(user_tfidf, tfidf_matrix)[0]))
    
    # Sắp xếp kết quả theo độ tương đồng giảm dần
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[:20]  # Lấy top 20 kết quả
    
    # Lấy thông tin cầu thủ từ data dựa trên chỉ số sim_scores
    recommendations = []
    for idx, score in sim_scores:
        if score > 0:  # Chỉ lấy kết quả có độ tương đồng lớn hơn 0
            player = data.iloc[idx]
            recommendations.append({
                "Tên": player['Tên'],
                "CLB": player['CLB'],
                "Quốc tịch": player['Quốc tịch'],
                "Tuổi": player['Tuổi'],
                "Chiều cao": player['Chiều cao'],
                "Cân nặng": player['Cân nặng'],
                "Bàn thắng": player['Bàn thắng'],
                 "Kiến tạo": player['Kiến tạo'],
                "Tắc bóng": player['Tắc bóng'],
                "Đánh chặn": player['Đánh chặn'],
                "Cứu thua": player['Cứu thua'],
                 "Thẻ vàng": player['Thẻ vàng'],
                 "Lạm phát": player['Lạm phát'],
                 "Phong cách và lối chơi": player['Phong cách và lối chơi'],
                 "Điểm mạnh": player['Điểm mạnh'],
                "Điểm yếu": player['Điểm yếu'],
                "Mô tả đặc điểm cá nhân": player['Mô tả đặc điểm cá nhân'],
                "Giá": player['Giá'],
                "Độ tương đồng": score
            })
    return recommendations
 
```
